```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carrega os dados
train = pd.read_csv("/workspaces/titanic-ml-project/data/train.csv")

# An√°lise de sobreviv√™ncia por diferentes caracter√≠sticas
print("üìä An√°lise de sobreviv√™ncia por caracter√≠sticas...")

plt.figure(figsize=(15, 10))

# Sobreviv√™ncia por sexo
plt.subplot(2, 3, 1)
sns.countplot(data=train, x='Sex', hue='Survived')
plt.title('Sobreviv√™ncia por Sexo')

# Sobreviv√™ncia por classe
plt.subplot(2, 3, 2)
sns.countplot(data=train, x='Pclass', hue='Survived')
plt.title('Sobreviv√™ncia por Classe')

# Sobreviv√™ncia por porto de embarque
plt.subplot(2, 3, 3)
sns.countplot(data=train, x='Embarked', hue='Survived')
plt.title('Sobreviv√™ncia por Porto de Embarque')

# Idade vs Sobreviv√™ncia
plt.subplot(2, 3, 4)
sns.boxplot(data=train, x='Survived', y='Age')
plt.title('Idade vs Sobreviv√™ncia')

# Tarifa vs Sobreviv√™ncia
plt.subplot(2, 3, 5)
sns.boxplot(data=train, x='Survived', y='Fare')
plt.title('Tarifa vs Sobreviv√™ncia')

# Irm√£os/C√¥njuges vs Sobreviv√™ncia
plt.subplot(2, 3, 6)
sns.countplot(data=train, x='SibSp', hue='Survived')
plt.title('Irm√£os/C√¥njuges vs Sobreviv√™ncia')

plt.tight_layout()
plt.show()

print("‚úÖ An√°lise completa realizada com sucesso!")
```
